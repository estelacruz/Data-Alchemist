**Task 1:** Use numpy to create a one-dimensional vector of length 40 with the numbers 1, 2, 3,
and 4 repeating 10 times in a cycle, that is [1 2 3 4 1 2 3 4 etc.]. Do not simply type all 40 numbers in. Rather, use broadcasting.



```python
import numpy as np 
array = np.array([1, 2, 3, 4])
repeats = np.tile(array, (10, 1))
repeats = np.concatenate(repeats)
repeats
```

**Task 3:**  Create an 8x8 numpy array such that it resembles the chess board pattern where you can assume that 0 means white and 1 means black. The lower right “square” should be

**Task 9:** Given the following dataframe, create a new dataframe that tells us the number of students enrolled in a particular course and sort them in a descending order based on number of students.


```python

students = pd.DataFrame(
{
"id": range(101, 111), "courses": [
          ["A", "B", "C"],
          ["A", "B", "E"],
          ["B", "D", "E"],
          ["C", "D", "F"],
          ["A", "D", "F"],
          ["B", "C", "E"],
          ["A", "D", "E"],
          ["C", "E", "F"],
          ["D", "E", "F"],
          ["B", "E", "F"],
], }
)
students

courses = {"A":0, "B":0, "C":0, "D":0,"E":0,"F":0}
stuC = students['courses']     
  
for x in stuC:
  #print(x) 
  for y in x:
    if(y in courses):
        courses[y] = courses.get(y) + 1
        #print(y, "has value ", courses[y])
    
    
#courses          
          
studentsEnrolled = pd.DataFrame(list(courses.items()), columns = ['Course', 'Enrolled'] )
studentsEnrolled.sort_values(by = ['Enrolled'], ascending = False)
```
